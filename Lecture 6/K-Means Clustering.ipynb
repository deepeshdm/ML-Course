```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
```

```python
path = r"C:\Users\dipesh\Desktop\Datasets\Mall_Customers.csv"

df = pd.read_csv(path)
df.head(10)
```

# K-Means using Elbow Method 

```python
path = r"C:\Users\dipesh\Desktop\Datasets\Mall_Customers.csv"

df = pd.read_csv(path)
df.head(10)
```

```python
df.rename(columns={"Spending Score (1-100)":"Spending Score"},inplace=True)

df.head(10)
```

```python
df = df[["Age","Spending Score"]]
df.head(10)
```

```python
K_Values = [2,3,4,5,6,7,8,9]
SSE = []

for k in K_Values:
    model = KMeans(n_clusters=k)
    model.fit(df)
    SSE.append(model.inertia_)

print(SSE)
```

```python
plt.plot(K_Values,SSE)
plt.xlabel("K")
plt.ylabel("SSE")
plt.title("ELBOW PLOT")
plt.show()
```

```python
model = KMeans(n_clusters = 4 )  # n_clusters = k
model.fit(df)
```

```python
cluster_labels = model.predict(df)
df["cluster"] = cluster_labels
```

```python
# scatterplot of clusters

plt.scatter(df["Age"],df["Spending Score"],c=df["cluster"])
plt.show()
```
