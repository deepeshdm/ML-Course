# ONE HOT ENCODING

```python
df = pd.read_csv(r"C:\Users\dipesh\Desktop\Iris.csv")
df.head(30)
```

```python
df = df.drop("Id",axis=1)
df.head()
```

```python
df.info()
```

```python
one_hot_species = pd.get_dummies(df["Species"])
one_hot_species
```

```python
df = pd.concat([df,one_hot_species],axis=1)
df.head(10)
```

```python
df = df.drop("Species",axis=1)
df.head(10)
```

# LABEL ENCODING

```python
df = pd.read_csv(r"C:\Users\dipesh\Desktop\Iris.csv")
df.head(30)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()

labels = encoder.fit_transform(df["Species"])
labels = pd.DataFrame(labels)
labels.columns = ["Labels"]
labels
```

```python
df = pd.concat([df,labels],axis=1)
df.head(10)
```

```python
df = df.drop(["Id","Species"],axis=1)
df.head(10)
```

```python
df = df.sample(frac=1)
df.head(20)
```
