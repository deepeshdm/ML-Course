```python
import numpy as np
import pandas as pd
from sklearn.ensemble import BaggingClassifier
```

```python
path = r"C:\Users\dipesh\Desktop\Datasets\diabetes.csv"

df = pd.read_csv(path)
df.head(10)
```

```python
x = df.drop("Outcome",axis=1)
x
```

```python
y = df["Outcome"]
y
```

```python
# data splitting

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_test
```

```python
y_test
```

```python
# Training the model

from sklearn.linear_model import LogisticRegression

model = BaggingClassifier(base_estimator= LogisticRegression(), n_estimators=10)
model.fit(x_train,y_train)
```

```python
# make predictions

i = model.predict_proba([[0,106,70,37,148,39.4,0.605,22]])
i
```

```python
# Finding the accuracy
accuracy = model.score(x_test,y_test)
accuracy
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
# Training the model

from sklearn.linear_model import LogisticRegression

model = AdaBoostClassifier(base_estimator= LogisticRegression(), n_estimators=5)
model.fit(x_train,y_train)
```
